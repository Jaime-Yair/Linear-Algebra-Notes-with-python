```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
```

```python
vals = np.linspace(-5,5,1000)
plt.figure(figsize=(5,5))

#The equations in y
eq0 = lambda x1: (1-x1)/-2 # x-2y = 1
y0 = eq0(vals)

eq1 = lambda x1: (11-3*x1)/2 # 3x +2y =11
y1 = eq1(vals)

plt.plot(vals,y0,color="blue")
plt.plot(vals,y1,color="red")
interception = np.argwhere(np.diff(np.sign(y0 - y1))).flatten() # finds the intersection between two graphs
plt.plot(vals[interception],y0[interception],'ko')
plt.legend(["3x +2y =11 ","x-2y = 1",str(np.round(vals[interception],2)) + str(np.round(y0[interception],2))]);

plt.xlim(-1,5)
plt.ylim(-1,5)
plt.axvline(x=0, linestyle='--')
plt.axhline(y=0, linestyle='--')
```

## 2.3 Elimination using matrices

```python
A=np.matrix([[2,4,-2],
            [4,9,-3],
            [-2,-3,7]])
```

```python
b = np.array([[2,8,10]])
```

```python
A_b=np.concatenate((A,b.T), axis=1)
A_b
```

```python
#Step 1

E1 = np.identity(3)
E1[1,0]=-2
```

```python
A_b1 = np.matmul(E1,A_b)
A_b1
```

```python
# Step 2

E2 = np.identity(3)
E2[2,0]=1
E2
```

```python
A_b2 = np.matmul(E2,A_b1)
A_b2
```

```python
# Step 3

E3 = np.identity(3)
E3[2,1]=-1
E3
```

```python
A_b3 = np.matmul(E3,A_b2)
A_b3
```

```python
# Doing everthing at once
np.matmul(E3,np.matmul(E2,np.matmul(E1,A_b)))
```

```python
# Solving the system of equations

x = np.linalg.solve(A,b.T)
x
```

```python
np.dot(A[0,:],x), np.dot(A[1,:],x), np.dot(A[2,:],x)  # The row picture
```

```python
float(x[0])*A[:,0]+float(x[1])*A[:,1]+float(x[2])*A[:,2]  # The column picture
```

##### The laws of matrix operations

```python
A,B,C=np.random.randint(1,5, size=(3,3)), np.random.randint(1,5, size=(3,3)), np.random.randint(1,5, size=(3,3))
c = 3
```

```python
A+B == B+A   # conmutative law
```

```python
c*(A+B) == c*A + c*B   # Distributive law
```

```python
A+(B+C) == (A+B)+C   # Associative law
```

```python
np.matmul(A,B) == np.matmul(B,A)  # The commutative law is usually broken
```

```python
np.matmul(A,B+C) == np.matmul(A,B)+np.matmul(A,C)  # distributive law from the left
```

```python
np.matmul(A+B,C) == np.matmul(A,C)+np.matmul(B,C)  # distributive law from the rithg
```

### 2.6 Elimination = Factorization: A = LU

```python
A = np.array([[2,1],[6,8]])
A
```

```python
E_21 = np.array([[1,0],[-3,1]])
```

```python
U = np.matmul(E_21, A) # forward from A to U
U
```

```python
E_21_inv = np.linalg.inv(E_21)
E_21_inv
```

```python
np.matmul(E_21_inv, U)  # back from U to A   LU=A
```

```python
E_32 = np.array([[1,0,0],
                [0,1,0],
                [0,-2/3,1]])
```

## 2.7 Transposes and permutation

```python
A.T ,np.transpose(A) 
```

```python
A.T==np.transpose(A) 
```

```python
# Transpose rules
```

```python
(A+B).T == A.T + B.T  # sum
```

```python
np.matmul(A,B).T == np.matmul(B.T,A.T) # Product
```

```python
np.linalg.inv(B).T == np.linalg.inv(B.T) # Inverse
```
